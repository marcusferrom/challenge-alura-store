### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento



```python
## faturamento loja 1
faturamento_loja1 = loja["Preço"].sum()
faturamento_loja1
```

```python
## faturamento das lojas restantes
faturamento_loja2 = loja2["Preço"].sum()
faturamento_loja3 = loja3["Preço"].sum()
faturamento_loja4 = loja4["Preço"].sum()

faturamento_loja2, faturamento_loja3, faturamento_loja4
```

```python
print(f"Loja 1: R$ {faturamento_loja1:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))
print(f"Loja 2: R$ {faturamento_loja2:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))
print(f"Loja 3: R$ {faturamento_loja3:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))
print(f"Loja 4: R$ {faturamento_loja4:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))
```

# 3. Média de Avaliação das Lojas


```python
media_avaliacao_loja1 = round(loja["Avaliação da compra"].mean(), 2)
media_avaliacao_loja2 = round(loja2["Avaliação da compra"].mean(), 2)
media_avaliacao_loja3 = round(loja3["Avaliação da compra"].mean(), 2)
media_avaliacao_loja4 = round(loja4["Avaliação da compra"].mean(), 2)

print("Loja 1:", media_avaliacao_loja1)
print("Loja 2:", media_avaliacao_loja2)
print("Loja 3:", media_avaliacao_loja3)
print("Loja 4:", media_avaliacao_loja4)
```

**Embora a diferença entre elas seja pequena, as médias de avaliação mostram que todas as lojas apresentam níveis semelhantes de satisfação.**

# 4. Produtos Mais e Menos Vendidos

```python
produtos_loja1 = loja["Produto"].value_counts()

# Maior quantidade
max_vendas1 = produtos_loja1.max()
mais_vendidos1 = produtos_loja1[produtos_loja1 == max_vendas1]

# Menor quantidade
min_vendas1 = produtos_loja1.min()
menos_vendidos1 = produtos_loja1[produtos_loja1 == min_vendas1]

print("Produto(s) mais vendido(s) - Loja 1:")
for produto, qtd in mais_vendidos1.items():
    print(f"{produto} ({qtd} vendas)")

print("\nProduto(s) menos vendido(s) - Loja 1:")
for produto, qtd in menos_vendidos1.items():
    print(f"{produto} ({qtd} vendas)")
```

```python
produtos_loja2 = loja2["Produto"].value_counts()

max_vendas2 = produtos_loja2.max()
mais_vendidos2 = produtos_loja2[produtos_loja2 == max_vendas2]

min_vendas2 = produtos_loja2.min()
menos_vendidos2 = produtos_loja2[produtos_loja2 == min_vendas2]

print("Produto(s) mais vendido(s) - Loja 2:")
for produto, qtd in mais_vendidos2.items():
    print(f"{produto} ({qtd} vendas)")

print("\nProduto(s) menos vendido(s) - Loja 2:")
for produto, qtd in menos_vendidos2.items():
    print(f"{produto} ({qtd} vendas)")
```

```python
produtos_loja3 = loja3["Produto"].value_counts()

max_vendas3 = produtos_loja3.max()
mais_vendidos3 = produtos_loja3[produtos_loja3 == max_vendas3]

min_vendas3 = produtos_loja3.min()
menos_vendidos3 = produtos_loja3[produtos_loja3 == min_vendas3]

print("Produto(s) mais vendido(s) - Loja 3:")
for produto, qtd in mais_vendidos3.items():
    print(f"{produto} ({qtd} vendas)")

print("\nProduto(s) menos vendido(s) - Loja 3:")
for produto, qtd in menos_vendidos3.items():
    print(f"{produto} ({qtd} vendas)")
```

```python
produtos_loja4 = loja4["Produto"].value_counts()

max_vendas4 = produtos_loja4.max()
mais_vendidos4 = produtos_loja4[produtos_loja4 == max_vendas4]

min_vendas4 = produtos_loja4.min()
menos_vendidos4 = produtos_loja4[produtos_loja4 == min_vendas4]

print("Produto(s) mais vendido(s) - Loja 4:")
for produto, qtd in mais_vendidos4.items():
    print(f"{produto} ({qtd} vendas)")

print("\nProduto(s) menos vendido(s) - Loja 4:")
for produto, qtd in menos_vendidos4.items():
    print(f"{produto} ({qtd} vendas)")
```

# 5. Frete Médio por Loja

```python
frete_medio_loja1 = round(loja["Frete"].mean(), 2)
frete_medio_loja2 = round(loja2["Frete"].mean(), 2)
frete_medio_loja3 = round(loja3["Frete"].mean(), 2)
frete_medio_loja4 = round(loja4["Frete"].mean(), 2)

print("Frete médio Loja 1: R$", frete_medio_loja1)
print("Frete médio Loja 2: R$", frete_medio_loja2)
print("Frete médio Loja 3: R$", frete_medio_loja3)
print("Frete médio Loja 4: R$", frete_medio_loja4)
```

## 6. Gráficos

```python
## gráfico 1 - faturamento por loja

import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

lojas = ["Loja 1", "Loja 2", "Loja 3", "Loja 4"]
faturamentos = [faturamento_loja1, faturamento_loja2, faturamento_loja3, faturamento_loja4]

plt.figure()
plt.bar(lojas, faturamentos)

plt.title("Faturamento por Loja")
plt.xlabel("Lojas")
plt.ylabel("Faturamento (R$)")

# formatar eixo Y para R$
def formatar_moeda(x, pos):
    return f'R$ {x:,.0f}'.replace(",", "X").replace(".", ",").replace("X", ".")

plt.gca().yaxis.set_major_formatter(FuncFormatter(formatar_moeda))

plt.show()
```

```python
# gráfico 2 - média de avaliação

avaliacoes = [media_avaliacao_loja1, media_avaliacao_loja2, media_avaliacao_loja3, media_avaliacao_loja4]

plt.figure()
plt.barh(lojas, avaliacoes)

plt.title("Média de Avaliação por Loja")
plt.xlabel("Média de Avaliação")
plt.ylabel("Lojas")

plt.show()
```

```python
# gráfico 3 - frete médio x avaliação

fretes = [frete_medio_loja1, frete_medio_loja2, frete_medio_loja3, frete_medio_loja4]

plt.figure()
plt.scatter(fretes, avaliacoes)

for i, loja in enumerate(lojas):
    plt.text(fretes[i], avaliacoes[i], loja)

plt.title("Frete Médio vs Avaliação")
plt.xlabel("Frete Médio (R$)")
plt.ylabel("Média de Avaliação")

plt.show()
```

## 7. Relatório

O objetivo desta análise foi avaliar o desempenho das quatro lojas do Senhor João para identificar qual delas deveria ser vendida, considerando os dados disponíveis para as 4 lojas. A partir dos dados analisados e dos gráficos, foi possível comparar o desempenho financeiro e operacional de cada unidade, buscando entender não apenas qual fatura mais, mas qual apresenta menor potencial e indicada para venda.

Em relação ao faturamento total, a Loja 1 se destacou como a que mais gera receita, enquanto a Loja 4 apresentou o menor faturamento entre as quatro. As categorias e os produtos mais vendidos mostraram padrões semelhantes entre as lojas, sem uma discrepância crítica que justificasse a venda com base apenas no mix de produtos. Sendo assim, com base apenas neste indicador não seria possível uma tomada de decisão segura, visto que as diferenças são mais de desempenho e não algo crítico.

Sobre a média das avaliações, todas as lojas ficaram muito próximas da nota 4,0, indicando boa percepção dos clientes. No entanto, a Loja 1 apresentou a menor média de avaliação, além de ter o frete médio mais alto entre as unidades. Já a Loja 3 teve a melhor avaliação com frete intermediário, mostrando melhor equilíbrio entre custo e experiência do cliente. A Loja 4, apesar de ter o menor faturamento, mantém avaliação estável e frete mais baixo, o que pode indicar potencial de crescimento caso haja ajustes estratégicos. Sendo assim, os indicadores passam a apontar mais divergências, que podem facilitar a decisão final.

Considerando todos os fatores, a recomendação é que o Senhor João venda a Loja 1. Apesar de ser a que mais fatura atualmente, ela combina o frete mais alto com a pior avaliação média, o que pode representar risco de queda de desempenho no longo prazo. As demais lojas demonstram maior equilíbrio entre receita, satisfação e competitividade. Por ser menos sólida e representar maior risco a longo prazo, a Loja 1 é a unidade com maior necessidade de ajustes estruturais e, portanto, a melhor candidata para venda.
